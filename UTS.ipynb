```python
import pandas as pd

df = pd.read_excel('/content/Data_BPS.xlsx')

selected_columns = ['JUMLAH PENDUDUK', 'LUAS WILAYAH (KM)', 'JUMLAH RUMAH', 'TEMBOK', 'JUMLAH SEKOLAH\n(SMP-SMA)', 'JUMLAH RS', 'Usia 15-64 (%)', 'Kompetitor Indihome', 'Kompetitior Lokal']
new_df = df[selected_columns]

new_df['JUMLAH RS'] = pd.to_numeric(new_df['JUMLAH RS'], errors='coerce')

new_df['JUMLAH RS'] = new_df['JUMLAH RS'].fillna(new_df['JUMLAH RS'].mean())

new_df['Kompetitor Indihome'] = new_df['Kompetitor Indihome'].apply(lambda x: 1 if x == 'ada' else 0)
new_df['Kompetitior Lokal'] = new_df['Kompetitior Lokal'].apply(lambda x: 1 if x == 'ada' else 0)

print(new_df.head())
```

```python
#Histogram Univariate
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

for column in selected_columns:
    plt.figure(figsize=(8, 6))

    if column in ['Kompetitor Indihome', 'Kompetitior Lokal']:
        sns.histplot(new_df[column], bins=[-0.5, 0.5, 1.5], discrete=True)
        plt.xticks([0, 1])
    else:
        sns.histplot(new_df[column], kde=True)

    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Descriptive Statistics dan Box Plot Univariate
for column in selected_columns:
  print(f"Descriptive Statistics for {column}:")
  print(new_df[column].describe())
  print("\n")

for column in selected_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=new_df[column])
    plt.title(f'Box Plot of {column}')
    plt.show()
```

```python
#Correlation Matrix Multivariate
correlation_matrix = new_df.corr()

print(correlation_matrix)

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#PCA Multivariate
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

features = ['JUMLAH PENDUDUK', 'LUAS WILAYAH (KM)', 'JUMLAH RUMAH', 'TEMBOK',
            'JUMLAH SEKOLAH\n(SMP-SMA)', 'JUMLAH RS', 'Usia 15-64 (%)',
            'Kompetitor Indihome', 'Kompetitior Lokal']
x = new_df[features].values

x = StandardScaler().fit_transform(x)

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data=principalComponents, columns=['Principal Component 1', 'Principal Component 2'])

finalDf = pd.concat([principalDf, new_df.reset_index(drop=True)], axis=1)

plt.figure(figsize=(8, 6))
plt.scatter(finalDf['Principal Component 1'], finalDf['Principal Component 2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Data')

explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio:", explained_variance)
plt.show()
```

```python
#Clustering Analysis Multivariate
from sklearn.cluster import KMeans
import numpy as np

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

optimal_k = 5

kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(x)

finalDf['Cluster'] = y_kmeans

plt.figure(figsize=(8, 6))
for i in range(optimal_k):
    plt.scatter(finalDf[finalDf['Cluster'] == i]['Principal Component 1'],
                finalDf[finalDf['Cluster'] == i]['Principal Component 2'],
                label=f'Cluster {i}')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clusters of Data')
plt.legend()
plt.show()

for i in range(optimal_k):
  print(f"Cluster {i} characteristics:")
  print(finalDf[finalDf['Cluster'] == i][features].describe())
  print("\n")
```
